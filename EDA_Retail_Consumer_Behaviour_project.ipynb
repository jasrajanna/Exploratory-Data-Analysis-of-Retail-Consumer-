<h1>Data Cleaning & Preprocessing</h1>

<h4>Removing duplicate entries based on Transaction ID </h4>

```python
df.drop_duplicates(subset='Transaction_ID',keep='first', inplace=True)
```

<h4>Dropping irrelevant columns from the dataset</h4>

```python

df.drop(['Transaction_ID','Name','Email','Address','Time','Zipcode','products','Phone','City','State'],axis=1,inplace=True)
```

<h4>Removing rows that contain null values</h4>

```python
df = df.dropna(subset=['Customer_ID']).reset_index(drop=True)

df = df.dropna(subset=['Date']).reset_index(drop=True)
```

<h4>Converting columns to proper data types</h4>

```python
if df['Customer_ID'].dtype == 'float64':
    df['Customer_ID'] = df['Customer_ID'].astype(int)
df['Customer_ID'] = df['Customer_ID'].astype(str)

df['Date'] = pd.to_datetime(df['Date'], format="%m/%d/%Y")
```

<h4>Handling Missing Values</h4>

```python
# Checking for number of null values
df.isnull().sum()
```

```python
# Extracting Month and Year from the Date column and handling missing values
df['Month']=df['Date'].dt.month_name()  
df['Month']=df['Month'].fillna(df['Month'])

df['Year']=df['Date'].dt.year
df['Year']=df['Year'].fillna(df['Year'])
df['Year']=df['Year'].astype(int)
```

```python
# Filling missing values in numerical columns using mean() and median()
df['Amount']=df['Amount'].fillna(df['Amount'].mean())

df['Total_Amount']=df['Total_Amount'].fillna(df['Total_Purchases']*df['Amount'])

df['Age']=df['Age'].fillna(df['Age'].median())
df['Age']=df['Age'].astype(int)

df['Total_Purchases']=df['Total_Purchases'].fillna(1)
```

```python
# Filling missing values in categorical columns using mode 
cat_cols = ['Customer_Segment',
            'Ratings',
            'Feedback',
            'Shipping_Method',
            'Payment_Method',
            'Order_Status',
            'Income',
            'Gender',
            'Country',
            'Product_Category',
            'Product_Brand',
            'Product_Type']

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

<h4>Resetting Index</h4>

```python
df['Index'] = range(1, len(df) + 1)
df = df.set_index('Index')
```

<h2>ðŸ’ Distribution of Revenue by Product Category using Boxplot</h2>

<h4>Goal</h4>
To compare the revenue distribution across different product categories after removing extreme outliers.

```python
# Visualizing the Revenue Distribution Across Product Categories (Violin Plot)
Q1 = df['Total_Amount'].quantile(0.25)
Q3 = df['Total_Amount'].quantile(0.75)
IQR = Q3 - Q1

df_rem_otlrs = df[
    (df['Total_Amount'] >= Q1 - 1.5 * IQR) &                    # filtering data to remove  outlier for boxplot
    (df['Total_Amount'] <= Q3 + 1.5 * IQR)
]

sns.boxplot(x='Product_Category',                                # Creating a violin plot to show the of revenue values of each product category          
               y='Total_Amount',
               data=df_rem_otlrs,
               color='orange',
               
              )
plt.title('Revenue Distribution Across Product Categories',       # Adding Title to viloin plot
          fontsize=13,color='maroon',fontweight='bold')
plt.xlabel('Product Category',fontsize=12,color='maroon')         # Labelling x-axis as Product Category
plt.ylabel('Revenue',fontsize=12,color='maroon')                  # Labelling y-axis as Revenue
plt.show()
```

<h4>Insights</h4>

* Revenue distribution varies across product categories
* Electronics has a higher median revenue, indicating stronger average sales performance.
* After removing extreme values, revenue comparisons across categories like Grocery  become more balanced and interpretable.

<h3 style="color:red;">Multivariate Analysis</h3>

<h2>ðŸ’ 2024- Month-wise Revenue Comparison Across Product Categories (Heatmap)</h2>

<h4>Goal</h4>
To compare how total revenue from each product category changes across january,febraury in 2024 and identify trends or category growth patterns.

```python
# Visualizing Product Category-wise Revenue Across Years
plt.figure(figsize=(12,6))
df_filtered = df[
    (df['Year'] == 2024) & (df['Month'].isin(['January', 'February'])) ]

sales_pivot = df_filtered.pivot_table(                  # Creating a Pivot Table for Year-wise Revenue by Product Category  
    values='Total_Amount',                     # Value to Aggregiate
    index='Product_Category',                  # Rows of Heatmap
    columns=['Year','Month'],                            # Columns of heatmap
    aggfunc='sum',                             # Summing Revenue
    fill_value=0                               # Filling missing value with zero
)
sns.heatmap(sales_pivot,                       # creating heatmap
            annot=True,                        # Showing values in cells
            cmap='coolwarm'                    # Giving colour for each cell
           )
plt.title('Product Category-wise Revenue Across Years',fontsize=14,color='g',fontweight='bold')
plt.xlabel('Year-Month',fontsize=12,color='g')
plt.ylabel('Product Category',fontsize=12,color='g')
plt.show()                                     
```

<h4>Insights</h4>

* Revenue contribution from each product category varies significantly across years, showing clear growth or decline patterns.
* Categories with darker (Electronics=8.1e+6) cells indicate strong year-wise performance and increasing sales momentum.
* Lighter cells reveal weaker revenue(Home decor=6.1e+6), highlighting periods of low category demand.
* Consistently high values across years indicate stable and high-performing categories in the market.
