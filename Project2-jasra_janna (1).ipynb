<h1>Data Cleaning</h1>

```python
# Checking for number of null values
df.isnull().sum()
```

```python
# Checking for duplicate rows in the dataset
df.duplicated().sum()
```

```python
# Removing duplicate entries 
df=df.drop_duplicates()
```

```python
# Dropping irrelevant columns from the dataset
df.drop(['Transaction_ID','Name','Email','Address','Time','Zipcode','products','Phone','City','State'],axis=1,inplace=True)
```

```python
# Removing rows that contain null values
df = df.dropna(subset=['Customer_ID']).reset_index(drop=True)

df = df.dropna(subset=['Date']).reset_index(drop=True)
```

```python
# Converting columns to proper data types
if df['Customer_ID'].dtype == 'float64':
    df['Customer_ID'] = df['Customer_ID'].astype(int)
df['Customer_ID'] = df['Customer_ID'].astype(str)

df['Date'] = pd.to_datetime(df['Date'], format="%m/%d/%Y")
```

```python
# Extracting Month and Year from the Date column and handling missing values
df['Month']=df['Date'].dt.month_name()  
df['Month']=df['Month'].fillna(df['Month'])

df['year']=df['Date'].dt.year
df['Year']=df['Year'].fillna(df['year'])
df['Year']=df['Year'].astype(int)
```

```python
# Filling missing values in numerical columns using mean() and median()
num_cols = ['Total_Amount','Amount']

for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

df['Age']=df['Age'].fillna(df['Age'].median())
df['Age']=df['Age'].astype(int)

df['Total_Purchases']=df['Total_Purchases'].fillna(1)
```

```python
# Filling missing values in categorical columns using mode 
cat_cols = ['Customer_Segment','Ratings','Feedback','Shipping_Method','Payment_Method','Order_Status','Income']

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# Filling missing values in object columns with 'Unknown'
obj_col=['Gender','Country','Product_Category','Product_Brand','Product_Type']
for col in obj_col:
    if df[col].isnull().any():
         df[col]=df[col].fillna("Unknown")
```

```python
df['Index'] = range(1, len(df) + 1)
df = df.set_index('Index')
```

<h2>💠Year-wise Revenue Comparison Across Product Categories (Heatmap)</h2>

<h4>Goal</h4>
To compare how total revenue from each product category changes across different years and identify trends or category growth patterns.
<h4>EDA Type</h4>
Multivariate Analysis
<h4>Method</h4>
Created a pivot table summarizing total revenue by product category for each year, then visualized it using a heatmap with annotated values for clear comparison.

```python
# Visualizing Product Category-wise Revenue Across Years
sales_pivot = df.pivot_table(                  # Creating a Pivot Table for Year-wise Revenue by Product Category  
    values='Total_Amount',                     # Value to Aggregiate
    index='Product_Category',                  # Rows of Heatmap
    columns='Year',                            # Columns of heatmap
    aggfunc='sum',                             # Summing Revenue
    fill_value=0                               # Filling missing value with zero
)
sns.heatmap(sales_pivot,                       # creating heatmap
            annot=True,                        # Showing values in cells
            cmap='coolwarm'                    # Giving colour for each cell
           )
plt.title('Product Category-wise Revenue Across Years',fontsize=14,color='g',fontweight='bold')
plt.xlabel('Year',fontsize=12,color='g')
plt.ylabel('Product Category',fontsize=12,color='g')
plt.show()                                     
```

<h4>Insights</h4>

* Revenue contribution from each product category varies significantly across years, showing clear growth or decline patterns.
* Categories with darker (high-value) cells indicate strong year-wise performance and increasing sales momentum.
* Lighter cells reveal weaker revenue years, highlighting periods of low category demand.
* Consistently high values across years indicate stable and high-performing categories in the market.
